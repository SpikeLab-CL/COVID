```python
import pandas as pd
import io
import requests
from matplotlib import pyplot as plt
```

```python
def read_url(url):
    content_raw = requests.get(url).content
    return pd.read_csv(io.StringIO(content_raw.decode('utf-8')))
```

# Historico de defunciones 

```python
url = "https://raw.githubusercontent.com/MinCiencia/Datos-COVID19/master/output/producto32/Defunciones_std.csv"

muertes = read_url(url)
muertes = muertes.rename(columns=lambda x: x.lower().replace(' ', '_'))
muertes['fecha'] = pd.to_datetime(muertes['fecha'])
muertes.head()
```

```python
codigo_region = (muertes[['region', 'codigo_region']]
                 .drop_duplicates()
                 .reset_index()
                 .drop(['index'], axis=1)
                 .sort_values(by='codigo_region'))
```

```python
codigo_region
```

```python
ultima_fecha = muertes.fecha.max()
ultima_fecha_shift = ultima_fecha.replace(year=2019)

date_range_19_20 = pd.DataFrame({'fecha': pd.date_range('2019-1-1', ultima_fecha)})
date_range_18_19 = pd.DataFrame({'fecha': pd.date_range('2018-1-1', ultima_fecha_shift)})

input_causal_effect = {}
for region in codigo_region.codigo_region.values:
    
    res = (muertes.codigo_region == region) & (muertes.fecha.dt.year >= 2019)
    muertes_region_19_20 = (muertes.loc[res].groupby(['fecha']) #se agrupa para sacar datos por region
                            .defunciones.sum()
                            .to_frame().reset_index())
    muertes_region_19_20 = pd.merge(muertes_region_19_20, date_range_19_20, on='fecha', how='right').fillna(0)
    muertes_region_19_20 = muertes_region_19_20.sort_values(by='fecha').reset_index().drop(['index'], axis=1)
    muertes_region_19_20['defunciones'] = muertes_region_19_20['defunciones'].astype('int')
    
    res = (muertes.fecha.dt.year >= 2018) & (muertes.fecha <= ultima_fecha_shift)
    muertes_regresores = (muertes.loc[res].groupby(['fecha', 'codigo_region']) #se agrupa para sacar datos por region
                          .defunciones.sum()
                           .to_frame().reset_index())

    muertes_regresores = pd.pivot_table(muertes_regresores, index='fecha', columns='codigo_region', values='defunciones')
    muertes_regresores = pd.merge(muertes_regresores.reset_index(), date_range_18_19, on='fecha', how='right').fillna(0)
    muertes_regresores.rename(columns={'fecha': 'fecha_18_19'}, inplace=True)
    
    df = pd.concat([muertes_region_19_20, muertes_regresores], axis=1)
    df = (df.rename(columns={'defunciones': 'y'})
      .rename(columns={i: f'x{i-1}' for i in range(1, 17)}))
    df['region'] = codigo_region.query(f'codigo_region == {region}').region.values[0]
    cols_to_use = ['y'] + [f'x{i}' for i in range(0, 16)] + ['fecha', 'region'] 
    df = df[cols_to_use]
    df.to_feather(f"../datos/input_modelo_{region}.feather")
```

```python
df = (df.rename(columns={'defunciones': 'y'})
      .rename(columns={i: f'x{i-1}' for i in range(1, 17)}))
df['region'] = codigo_region.query(f'codigo_region == {region}').region.values[0]
cols_to_use = ['y'] + [f'x{i}' for i in range(0, 16)] + ['fecha', 'region'] 
df = df[cols_to_use]
```
