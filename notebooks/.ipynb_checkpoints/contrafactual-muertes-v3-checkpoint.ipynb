```python
import pandas as pd
from fbprophet import Prophet
from sklearn.preprocessing import MinMaxScaler
from tqdm.notebook import tqdm
from multiprocessing import Pool, cpu_count
from functools import partial
import numpy as np
import plotly.express as px
from matplotlib import pyplot as plt
import io
import requests

def read_url(url):
    content_raw = requests.get(url).content
    return pd.read_csv(io.StringIO(content_raw.decode('utf-8')))

PID = 'covid-spike'
tqdm.pandas()
```

# Cargar datos

## Defunciones totales

```python
sql = """
SELECT Region,
       Fecha,
       SUM(Defunciones) AS defunciones
FROM reportes_minsal.Defunciones_std
GROUP BY 1, 2
"""
muertes = pd.read_gbq(sql, project_id=PID,use_bqstorage_api=True)
muertes['Fecha'] = pd.to_datetime(muertes['Fecha'])
muertes.sort_values(by=['Region', 'Fecha'], inplace=True)

replace_region = {'Aysén del General Carlos Ibáñez del Campo': 'Aysén',
                  'La Araucanía': 'Araucanía',
                  "Libertador General Bernardo O'Higgins": 'O’Higgins',
                  'Magallanes y de la Antártica Chilena': 'Magallanes',
                  'Metropolitana de Santiago': 'Metropolitana'}

muertes['Region'] = muertes.Region.replace(replace_region)
muertes.sort_values(by=['Region', 'Fecha'], inplace=True)

muertes.head()
```

## Defunciones por COVID

```python
sql = """
SELECT Region,
       Fecha,
       Total AS defunciones_covid_acumuladas,
       Total - IF(LAG(Total) OVER 
       (PARTITION BY Region ORDER BY Fecha) IS NULL, 0, LAG(Total) 
       OVER (PARTITION BY Region ORDER BY Fecha)) AS defunciones_covid
FROM reportes_minsal.FallecidosCumulativo_std
"""
muertes_covid = pd.read_gbq(sql, project_id=PID,use_bqstorage_api=True)
muertes_covid['Fecha'] = pd.to_datetime(muertes_covid['Fecha'])
```

```python
url = "https://raw.githubusercontent.com/MinCiencia/Datos-COVID19/master/output/producto37/Defunciones_std.csv"
muertes_corregidas = read_url(url)
muertes_corregidas['Fecha'] = pd.to_datetime(muertes_corregidas['Fecha'])
muertes_corregidas.sort_values(by=['Fecha', 'Publicacion'], inplace=True)
muertes_corregidas = (muertes_corregidas
                      .groupby("Fecha").Total.last()
                      .to_frame('defunciones_covid_v2').reset_index())
date_range = pd.date_range(muertes_corregidas.Fecha.min(), muertes_corregidas.Fecha.max())
date_range = pd.DataFrame({'Fecha': date_range})
muertes_corregidas = pd.merge(muertes_corregidas, date_range, on='Fecha', how='right').fillna(0)
muertes_corregidas['defunciones_covid_v2'] = muertes_corregidas['defunciones_covid_v2'].astype(int)
muertes_corregidas.sort_values(by='Fecha', inplace=True)
muertes_corregidas['defunciones_covid_v2_acc'] = np.cumsum(muertes_corregidas['defunciones_covid_v2'])
```

# Modelamiento

Entrenamos un modelo de prophet con datos de muertes por región hasta el 1 de marzo y usando sumas moviles de 7 dias

```python
muertes.head()
```

```python
muertes
```

```python
muertes_totales = (muertes.groupby('Fecha')
                   .defunciones.sum()
                   .to_frame('defunciones')
                   .reset_index())
muertes_totales['log_defunciones'] = muertes_totales.eval('log(defunciones)')
```

```python
cut_date = pd.to_datetime('2020-3-1')

time_series = (muertes_totales
              .rename(columns={'Fecha': 'ds', 'log_defunciones': 'y'}))
time_series_cut = time_series[time_series.ds < cut_date]

model = Prophet(seasonality_mode='multiplicative',
                yearly_seasonality=True, 
                weekly_seasonality=True, 
                daily_seasonality=False)

model.fit(time_series_cut)
forecast = model.predict(time_series[['ds']])

cols = ['ds', 'yhat_lower', 'yhat_upper', 'yhat']
cols_log = [c for c in cols if c != 'ds']
forecast[cols_log] = np.exp(forecast[cols_log])

rename = {'yhat_lower': 'defunciones_predict_lower',
          'yhat_upper': 'defunciones_predict_upper',
          'yhat': 'defunciones_predict',
          'ds': 'Fecha'}
forecast = forecast[cols].rename(columns=rename)
muertes_totales = pd.merge(muertes_totales, forecast, on= ['Fecha'])
```

```python
fecha_from = pd.to_datetime('2020-1-1')
df_res = muertes_totales[muertes_totales.Fecha >= fecha_from].set_index('Fecha')

ax = (df_res
      .defunciones.plot(figsize=(15, 10), label='defunciones', style='o'))

(df_res
 .defunciones_predict.plot(ax=ax, style='-', label='predicción defunciones'))

ax.fill_between(df_res.index, 
             df_res['defunciones_predict_lower'], 
             df_res['defunciones_predict_upper'], 
             color='orange',
             alpha=0.5)

ax.legend()
```

```python
for region in muertes.Region.unique():
    cols = ['defunciones_msum7dias', 'defunciones_msum7dias_predict_lower', 
            'defunciones_msum7dias_predict_upper', 'defunciones_msum7dias_predict']
    muertes_region = muertes.query(f'Region == "{region}"').set_index('Fecha')[cols]

    ax = muertes_region['defunciones_msum7dias'].plot(style='.', figsize=(10, 7))
    muertes_region['defunciones_msum7dias_predict'].plot(color='black', ax=ax)
    ax.set_title(region)
    plt.show()
```

```python
contrafactual = pd.merge(muertes_covid, muertes, on=['Region', 'Fecha'])
```

```python
contrafactual
```

```python
contrafactual['delta_defunciones_msum7dias'] = contrafactual.eval('(defunciones_msum7dias-defunciones_msum7dias_predict < 0)*0 + (defunciones_msum7dias-defunciones_msum7dias_predict >= 0)*(defunciones_msum7dias-defunciones_msum7dias_predict)')
contrafactual['delta_defunciones_msum7dias_lower'] = contrafactual.eval('(defunciones_msum7dias-defunciones_msum7dias_predict_upper < 0)*0 + (defunciones_msum7dias-defunciones_msum7dias_predict_upper >= 0)*(defunciones_msum7dias-defunciones_msum7dias_predict_upper)')
contrafactual['delta_defunciones_msum7dias_upper'] = contrafactual.eval('(defunciones_msum7dias-defunciones_msum7dias_predict_lower < 0)*0 + (defunciones_msum7dias-defunciones_msum7dias_predict_lower >= 0)*(defunciones_msum7dias-defunciones_msum7dias_predict_lower)')
```

```python
contrafactual.head()
```

```python
def get_acc(x):
    x_acc = x.copy()
    nshifts = len(x) // 7
    for k in range(1, nshifts+1):
        x_acc += x.shift(7*k).fillna(0)
    return x_acc
```

```python
contrafactual['delta_defunciones_acc'] = contrafactual.groupby('Region')['delta_defunciones_msum7dias'].apply(get_acc)
contrafactual['delta_defunciones_acc_lower'] = contrafactual.groupby('Region')['delta_defunciones_msum7dias_lower'].apply(get_acc)
contrafactual['delta_defunciones_acc_upper'] = contrafactual.groupby('Region')['delta_defunciones_msum7dias_upper'].apply(get_acc)
```

```python
cols_lines = ['defunciones_covid_acumuladas', 'delta_defunciones_acc']
for region in muertes.Region.unique():
    df_res = contrafactual.query(f"Region == '{region}'").set_index('Fecha')
    ax = df_res[cols_lines].plot(figsize=(10, 7))
    ax.fill_between(df_res.index, 
                 df_res['delta_defunciones_acc_lower'], 
                 df_res['delta_defunciones_acc_upper'], 
                 color='orange',
                 alpha=0.5)
    delta_upper = int(df_res[cols_lines].iloc[-1, 1] - df_res.iloc[-1]['delta_defunciones_acc_lower'])
    delta_lower = int(df_res[cols_lines].iloc[-1, 1] - df_res.iloc[-1]['delta_defunciones_acc_upper'])
    ax.set_title(region + f' delta in [{delta_lower}, {delta_upper}]')
```
