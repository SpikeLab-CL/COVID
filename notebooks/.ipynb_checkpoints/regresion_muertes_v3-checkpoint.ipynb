```python
import h2o
import pandas as pd
from h2o.estimators.gbm import H2OGradientBoostingEstimator
from h2o.automl import H2OAutoML
import numpy as np
import plotly.express as px
import sys
sys.path.append("../../SpikePy")
import SpikePy as sp
from h2o.estimators.gbm import H2OGradientBoostingEstimator
from matplotlib import pyplot as plt

H2O_server = h2o.init(port=54321, nthreads=-1)
h2o.remove_all() 

PID = 'covid-spike'
```

```python
sql = """
SELECT *
FROM reportes_minsal.consolidado_region
"""
df = pd.read_gbq(sql, 
                 project_id=PID, use_bqstorage_api=True)
df['fecha'] = pd.to_datetime(df['fecha'])
df.sort_values(by=['Region', 'fecha'], inplace=True)
```

```python
sql = """
SELECT * EXCEPT(nom_region, region),
       nom_region AS region
FROM censo_2017.poblacion_region_comuna
"""
censo = pd.read_gbq(sql, project_id=PID,use_bqstorage_api=True)
censo.rename(columns=lambda x: x.lower(), inplace=True)
censo.rename(columns= lambda x: x.replace('__', '_'), inplace=True)
replace_region = {'REGIÓN DE ARICA Y PARINACOTA': 'Arica y Parinacota',
                  'REGIÓN DE TARAPACÁ': 'Tarapacá',
                  'REGIÓN DE ANTOFAGASTA': 'Antofagasta',
                  'REGIÓN DE ATACAMA': 'Atacama',
                  'REGIÓN DE COQUIMBO': 'Coquimbo',
                  'REGIÓN DE VALPARAÍSO': 'Valparaíso',
                  'REGIÓN METROPOLITANA DE SANTIAGO': 'Metropolitana',
                  "REGIÓN DEL LIBERTADOR GENERAL BERNARDO O'HIGGINS": 'O’Higgins',
                  'REGIÓN DEL MAULE': 'Maule',
                  'REGIÓN DEL ÑUBLE': 'Ñuble',
                  'REGIÓN DEL BIOBÍO':'Biobío',
                  'REGIÓN DE LA ARAUCANÍA': 'Araucanía',
                  'REGIÓN DE LOS RÍOS': 'Los Ríos',
                  'REGIÓN DE LOS LAGOS': 'Los Lagos',
                  'REGIÓN DE AYSÉN DEL GENERAL CARLOS IBÁÑEZ DEL CAMPO': 'Aysén',
                  'REGIÓN DE MAGALLANES Y DE LA ANTÁRTICA CHILENA': 'Magallanes'}
censo['region'] = censo['region'].replace(replace_region)
```

```python
sum_cols = ['total_vivi',
            'particular',
            'colectivas',
            'hombres',
            'mujeres']

mean_cols = ['indice_mas', 'indice_dep', 'ind_dep_ju', 'ind_dep_ve']

censo_region = (pd.concat([censo.groupby('region')[sum_cols].sum(), censo.groupby('region')[mean_cols].mean()],
                  axis=1).reset_index())

sum_hombres_mujeres = censo_region[['hombres', 'mujeres']].sum(axis=1)
censo_region['hombres'] = censo_region['hombres']/sum_hombres_mujeres
censo_region['mujeres'] = censo_region['mujeres']/sum_hombres_mujeres
```

```python
cols_lag = {15: ['retail_and_recreation_percent_change_from_baseline',
                 'grocery_and_pharmacy_percent_change_from_baseline',
                 'parks_percent_change_from_baseline',
                 'transit_stations_percent_change_from_baseline',
                 'workplaces_percent_change_from_baseline',
                 'residential_percent_change_from_baseline',
                 'casos_totales_acum', 'test_positivo'],
            7:  ['CFR_9d', 
                 'ln_mediamovil_7d_muertes_acum', 
                 'var_percent_muertes_acum', 
                 'CAGR_muertes_acum_9d',
                 'casos_totales_acum'],
            8:  ['ln_mediamovil_7d_muertes_acum']
            }
```

```python
def add_lags(df, cols_lag):
    for lag, cols in cols_lag.items():
        df_lag = df.groupby('Region')[cols].shift(lag).rename(columns={c:c+f'_lag{lag}' for c in cols})
        df = pd.concat([df, df_lag], axis=1)        
    return df
```

```python
df = add_lags(df, cols_lag)
```

```python
columns = list(df.columns)
columns.sort()
columns
```

```python
regions_to_use = ['Metropolitana', 'Valparaíso', 'Araucanía', 'Antofagasta']
regions_to_use = ['Metropolitana']
```

```python
df = df.query("Region in @regions_to_use")
```

```python
fechas = df.fecha.unique()
fechas.sort()
days_passed = pd.DataFrame({'fecha': fechas, 'days_passed': range(len(fechas))})
df = pd.merge(df, days_passed, on='fecha', how='left')
```

```python
numeric = ['CAGR_muertes_acum_9d_lag7',
           'CFR_9d_lag7', 
           'casos_totales_acum_lag15',
           'casos_totales_acum_lag7',
           'grocery_and_pharmacy_percent_change_from_baseline_lag15', 
           'ln_mediamovil_7d_muertes_acum_lag7',
           'parks_percent_change_from_baseline_lag15', 
           'residential_percent_change_from_baseline_lag15',
           'retail_and_recreation_percent_change_from_baseline_lag15', 
           'test_positivo_lag15', 
           'transit_stations_percent_change_from_baseline_lag15', 
           'var_percent_muertes_acum_lag7', 
           'workplaces_percent_change_from_baseline_lag15', 
           'ln_mediamovil_7d_muertes_acum_lag8',
           'days_passed']

categs = ['Region']
categs = []

target = 'ln_mediamovil_7d_muertes_acum'

features = numeric + categs
```

```python
fechas = df.fecha.unique()
fechas_str = [np.datetime_as_string(f, unit='D') for f in fechas]
len(fechas)
```

```python
min_days_train = 20
test_days = 7

auto_ml = False
horas_automl = 0.01

varimp_list = []
results_list = []




for i, fecha in enumerate(fechas[min_days_train:-test_days]):
    df['is_test'] = np.NaN
    fecha_train_str = np.datetime_as_string(fecha, unit='D')
    df.loc[df.fecha == fecha + pd.Timedelta('7D'), 'is_test'] = 1
    df.loc[df.fecha <= fecha, 'is_test'] = 0 


    df_train = h2o.H2OFrame(df[df.is_test == 0])
    df_test = h2o.H2OFrame(df[df.is_test == 1])

    for c in numeric + [target]:
        df_train[c] = df_train[c].asnumeric()
        df_test[c] = df_test[c].asnumeric()

    for c in categs:
        df_train[c] = df_train[c].asfactor()
        df_test[c] = df_test[c].asfactor()

    if auto_ml:
        aml = H2OAutoML(max_runtime_secs = int(60*60*horas_automl), seed=1313, 
                        project_name=f'model_{i}', include_algos=['GBM'])
        aml.train(x=features, 
                  y=target,
                  training_frame=df_train, )
        lb = aml.leaderboard

        gbm_id = [m for m in lb['model_id'].as_data_frame().model_id.values if 'GBM' in m][0]
        gbm = h2o.get_model(gbm_id)
    else:
        params = {'max_depth': 16,
                  'ntrees': 78,
                  'min_rows': 5.0,
                  'learn_rate': 0.05,
                  'sample_rate': 0.8,
                  'col_sample_rate': 0.7,
                  'monotone_constraints': {'days_passed': 1}
                  }
        gbm = H2OGradientBoostingEstimator(nfolds=5, **params, model_id=f'model_{i}')
        gbm.train(x=features, y=target, training_frame=df_train)

    varimp = (pd.DataFrame(gbm.varimp())
                          .set_index(0)[2].to_frame()
                          .reset_index()
                          .rename(columns={0:'feature', 2: 'imp'}))

    varimp['fecha'] = fecha
    varimp_list += [varimp]

    results = df.loc[df.is_test == 1, ['fecha', 'Region', target]]
    results[target+'_predict'] = gbm.predict(df_test).as_data_frame().predict.values

    results[target.replace('ln_', '')] = np.exp(results[target]) - 1
    results[(target+'_predict').replace('ln_', '')] = np.exp(results[target+'_predict']) - 1

    results['muertes_prox_7d'] = 7 * (results[target.replace('ln_', '')] - np.exp(df.loc[df.fecha==fecha, target]).values) + 1
    results['muertes_prox_7_predict'] = 7 * (results[(target+'_predict').replace('ln_', '')] - np.exp(df.loc[df.fecha==fecha, target]).values) + 1
    
    results_list += [results]
```

```python
results = pd.concat(results_list)
varimp = pd.concat(varimp_list)
```

```python
sp.pdplot(varimp, variables=['feature', 'fecha'], target='imp', numeric=False, target_type='regression')
```
